## Interpolate the data


```python
start_time = pd.Timestamp('2021-08-01 00', tz = df_max_torque_3.index.tz)
end_time = pd.Timestamp('2022-12-31 00', tz = df_max_torque_3.index.tz)
step_seconds=10
chunk_len_min=60 
df_max_torque_1_interpolated = interpolate_chunks(df_max_torque_1, start_time, end_time, step_seconds, chunk_len_min)
df_max_torque_2_interpolated = interpolate_chunks(df_max_torque_2, start_time, end_time, step_seconds, chunk_len_min)
df_max_torque_3_interpolated = interpolate_chunks(df_max_torque_3, start_time, end_time, step_seconds, chunk_len_min)
df_max_torque_4_interpolated = interpolate_chunks(df_max_torque_4, start_time, end_time, step_seconds, chunk_len_min)
df_max_torque_5_interpolated = interpolate_chunks(df_max_torque_5, start_time, end_time, step_seconds, chunk_len_min)
df_max_torque_6_interpolated = interpolate_chunks(df_max_torque_6, start_time, end_time, step_seconds, chunk_len_min)
df_max_torque_7_interpolated = interpolate_chunks(df_max_torque_7, start_time, end_time, step_seconds, chunk_len_min)
```

## Merge all the axes in a single data frame


```python
dfs = [
    df_max_torque_1_interpolated,
    df_max_torque_2_interpolated,
    df_max_torque_3_interpolated,
    df_max_torque_4_interpolated,
    df_max_torque_5_interpolated,
    df_max_torque_6_interpolated,
    df_max_torque_7_interpolated
]
                                                                    
 # Rename columns to value_1, value_2, ..., value_7
for i, df in enumerate(dfs, start=1):
    df.columns = [f'value_{i}']

# Concatenate horizontally
merged_df = pd.concat(dfs, axis=1)
```

## delete missing rows

```python
merged_df = merged_df[merged_df.notna().all(axis=1)].copy()
merged_df
```

## Create case_id (break the data into multiple functions and enumerate)

```python
merged_df.loc[:, 'case_id'] = (np.arange(len(merged_df)) // 360 + 1).astype(int)
merged_df
```

## Set periodic time intervals for ease of analysis

```python
merged_df['dup_timestamp'] = merged_df.groupby('case_id', group_keys=False).apply(
    lambda g: start_time + (g.index - g.index.min())
).explode().values
```
